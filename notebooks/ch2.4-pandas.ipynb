## Pandas: Data Analysis

- Pandas contains data structures and data manipulation tools designed to make data cleaning
and analysis fast and easy in Python. 
- Pandas is often used in tandem with numerical omputing tools like NumPy and SciPy, analytical libraries like statsmodels and scikit-learn, and data visualization libraries like matplotlib.


- Fundemantal pandas data type is **series**. This is a one-dimensional array-like object containing a sequence of 
values of similar types to NumPy types and an associated array of data labels, called its index


- **DataFrame** is a 2-dimensional data structure that can store data of different types (including characters, integers, floating point values, categorical data and more) in columns.

- Dataframe is simply a container of for multiple 'series' objects.. **Each column in a DataFrame is a Series**

### Handling Missing Data

- This is a HUGE subject. Please have a look at pandas page [here](https://pandas.pydata.org/docs/user_guide/missing_data.html). You can also find this tutorial useful [here](https://www.youtube.com/watch?v=KdmPHEnPJPs&t=1s).


- Domain knowladge, target variable etc matters. 


- Here we will simply do the following; if the missing data is categorical, we will replace them with the category with highest count. If it is numerical, replace them with the median(average) of that column. 

```python
#check missing data counts
df.isnull().sum()
```

```python
#parent_education
#find the mode
mode_parent_edu = df['parent_education'].mode()[0]
print(mode_parent_edu)

df['parent_education'].fillna(mode_parent_edu,inplace=True)

#without "inplace"
#df['parent_education'] = df['parent_education'].fillna(mode_parent_edu,inplace=True)

#check
df['parent_education'].isnull().sum()
```

```python
#writing score
#mean
writing_avg = df['writing_score'].mean()
print(writing_avg)
#fill out NA with mean
df['writing_score'].fillna(writing_avg,inplace=True)
#check
df['writing_score'].isnull().sum()
```

### Further inspection of  Scores with Histogram, Box and Violin Plots

- Let's see how parent education is related to exam scores. We can try scatter plot but it will be too crowded. We will take advantage of three types of plot; histogram, violin plot and boxplot.


- We will also have another look at gender and race.

- A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins. It is way more useful than count plot for percentage.
- multiple: {“layer”, “dodge”, “stack”, “fill”} Approach to resolving multiple elements when semantic mapping creates subsets. Only relevant with univariate data. Partition into hues
- stat: Aggregate statistic to compute in each bin. percent: normalize such that bar heights sum to 100 in each category. Notice that the sum of each bar adds up to 100%. 
- shrink: number Scale the width of each bar relative to the binwidth by this factor. Only relevant with univariate data.
- common_norm{bool}: If True and using a normalized statistic, the normalization will apply over the full dataset. Otherwise, normalize each histogram independently.


```python
# we would like to have bars in every binrange/bins=(100-0)/5 = 20-point interval
plt.figure(figsize=(10,6))
sns.histplot(data = df,x = 'math_score',hue = 'parent_education',multiple = "dodge", 
             binrange = [0,100], bins = 5, shrink = 0.5)
```

- In a box plot the horizontal line indicates the median value. The range of the box captures
25% to 75% percentile ranges. The long-thin line-called whiskers- displays the data range. Box-plot also also shows the outliers.


- In a violin plot, vertical short-thick line captures 25% to 75% percentile ranges while long-thin line displays the data range. Little white dot is the median value.Violin plot outer curve is the probability density distribution(PDE). The violin plot does not display the outliers but it perfectly captures how the data is actually distributed across the data range. We dont see this in a box plot.



- For example, in the violin blot below, look how there are some students having parents with high school degree that got low scores in mathematics. The tail of the brown violin extends all the way down. However, this is not the case for students who have parents with master's degree. On average, they perform better than all others. Look at the median (white dot) in the green violin, it is above all others. The very same result is also visible from the subsequent box plot. 
Here is the solid results verifying this claim;

```python
math_parent = df.groupby(['parent_education'])['math_score']
math_parent.median()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(data = df, x = 'parent_education', y = 'math_score')
plt.grid(True)
```
